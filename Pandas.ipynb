# Grouping -. it is used to see the agrregated values

```python
df.groupby("mnth")["temp"].mean()
#mnth->grouping column
#temp->Aggregating column
#mean->aggregating funct
```

```python
df[df["mnth"]==1]["temp"].mean()
```

```python
df.groupby("mnth")[["temp","hum","windspeed"]].max()
```

```python
#temp --> max/min
#windspeed --> mean/median
#hum --> count/var/std/mean
#cnt --> std
```

```python
print(df.groupby("mnth")["temp"].max())
print(df.groupby("mnth")["temp"].min())
print(df.groupby("mnth")["windspeed"].mean())
print(df.groupby("mnth")["windspeed"].median())
print(df.groupby("mnth")["hum"].count())
print(df.groupby("mnth")["hum"].var())
print(df.groupby("mnth")["hum"].std())
print(df.groupby("mnth")["hum"].mean())
print(df.groupby("mnth")["cnt"].std())
```

```python
res={
    "temp":["max","min"],
    "windspeed":["mean","median"],
    "hum":["count","var","std","mean"],
    "cnt":["std"]
}
```

```python
df.groupby("mnth").agg(res)
```
